```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import zscore
```

```python
nbadata = pd.read_csv('NBA rookies.csv')
```

```python
nbadata
```

```python
nbadata.info()
```

Let's snag the stats that we need to power our scoring.

```python
keymetrics = nbadata[['Player','Tm','Season','Criteria TS%','Criteria USG%','Per Game TRB','Per Game AST',
                     'Per Game STL','Per Game BLK','Per Game PTS','Per Game TOV','Advanced PER',
                    'Advanced WS/48', 'Advanced BPM','Advanced VORP','Advanced FTr']].set_index(['Player','Tm','Season'])
```

Now, we'll get the zscore for these players for all the key stats.

```python
weighted = keymetrics.apply(zscore)
```

```python
weighted.info()
```

```python
weighted['Total Score'] = (1.25 * weighted['Criteria TS%'] + 1.25 * weighted['Criteria USG%'] + .75 * weighted['Per Game TRB'] + 
                           .75 * weighted['Per Game AST']  + .75 * weighted['Per Game TOV'] +
                           .75* weighted['Per Game STL'] + .75 * weighted['Per Game BLK'] 
                           + weighted['Per Game PTS'] + 1.25 * weighted['Advanced PER'] + 1.25 * weighted['Advanced WS/48'] + 1.25 * 
                           weighted['Advanced BPM']  + 1.25 * weighted['Advanced VORP'])
```

```python
weighted.sort_values('Total Score',ascending=False)
```

```python
totalscore = weighted.sort_values('Total Score',ascending=False)
```

```python
totalscore
```

```python
totalscore['Rank'] = totalscore['Total Score'].rank(pct=False,ascending=False)
```
